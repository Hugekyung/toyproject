```python
from collections import Counter
from konlpy.tag import Okt

okt = Okt()
noun = okt.nouns(review)
for i,v in enumerate(noun):
    if v == '리니지' or v == '게임' or len(v) < 2 or v == '엔씨':
        noun.pop(i)

count = Counter(noun)

noun_lst = count.most_common(10)
for i in noun_lst:
    print(i)
```

```python
top_words = noun_lst

top_words[0][1]
```

```python
import pandas as pd

df = pd.DataFrame(top_words, columns=['word', 'word_num'] )
lin_word = list(df['word'])
lin_word_num = list(df['word_num'])
```

# 브이포

```python
df_v4 = pd.read_csv(r'V4.csv')
com = list(df_v4['comment'])
review = ', '.join(com)
```

```python
from collections import Counter
from konlpy.tag import Okt

okt = Okt()
noun = okt.nouns(review)
for i,v in enumerate(noun):
    if v == '브이포' or v == '게임' or len(v) < 2 or v == '넥슨' or v == 'v4':
        noun.pop(i)

count = Counter(noun)

noun_lst = count.most_common(10)
for i in noun_lst:
    print(i)
```

```python
v4_top_words = noun_lst
```

```python
df_v4 = pd.DataFrame(v4_top_words, columns=['word', 'word_num'] )
word = list(df_v4['word'])
word_num = list(df_v4['word_num'])
```

# 종합

```python
plt.figure(figsize=(10,5))
sns.barplot(data=df_v4,x=word_num, y=word)
plt.title('브이포 리뷰 내 등장 단어 TOP 10')

plt.figure(figsize=(10,5))
sns.barplot(data=df,x=lin_word_num, y=lin_word)
plt.title('리니지 리뷰 내 등장 단어 TOP 10')
```

- 리니지와 브이포 모두 '과금'의 등장 빈도가 가장 높음. 유저들의 과금에 대한 의견이 많은 것으로 해석됨
- 브이포는 '과금', '현질'이라는 부정적인 단어가 상위에 있고, 리니지는 '과금', '현질', '작업장', '제한', '유도'라는 5개의 부정적인 단어가 포함되어 상대적으로 더 부정적인 단어가 자주 등장하고 있음. 특히 리니지는 1위인 '과금'이 압도적인 등장 수를 보여주고 있음.
- 브이포 등장 단어에는 리니지에는 없는 '그래픽'과 '업데이트'가 등장하며 과금 이외의 주제도 자주 등장하는 것을 확인할 수 있음.
